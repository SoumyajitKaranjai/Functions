1.What is the difference between a function and a method in Python?

**Functions:-**
* Defined independently, not within a class.  
*   Can be called by their name from anywhere in the code.


*   Do not require a self parameter.
*   Operate only on the data explicitly passed to them as arguments.

Example:

def add(x, y):
    
    return x + y

result = add(2, 3)#Here, add is a function that can be called directly.

**Method:-**



*   Defined inside a class.
*   Associated with an object (instance) or the class itself.


*   Called on an object or class using dot notation (e.g., object.method()).
*  Can access and modify the state (data) of the object they belong to.

Example:

class Calculator:
    def add(self, x, y):
        
        return x + y

calc = Calculator()

result = calc.add(2, 3)

Here, add is a method of the Calculator class and  is called on an instance of that class.

 **Arguments:-**

Example:

5.What are iterators in Python and how do they differ from iterables?

**Iterator:-**
* An iterator is an object that allows traversal through a countable number of values. It is used to iterate over iterable objects like lists, tuples, sets, etc.


*   It must implement two methods: __iter__() (which returns itself) and __next__() (which returns the next item or raises StopIteration when exhausted).
*  Iterators are used to fetch data from iterables one element at a time, maintaining internal state to remember where it is in the sequence.


*   Once an iterator is exhausted, it cannot be reset or reused; you need to create a new iterator from the iterable if you want to iterate again.

**Iterables:-**
*   An iterable is any Python object capable of returning its members one at a time, allowing it to be looped over.


*   Common examples include lists, tuples, strings, dictionaries, and sets.
*   An iterable must implement the __iter__() method, which returns an iterator object.


*   You can create an iterator from an iterable by passing it to the iter() function.

7. What are the advantages of using generators over regular functions?


*   Generators yield values one at a time and do not store the entire sequence in memory. This makes them ideal for processing large datasets or infinite sequences, as they only keep the current state, significantly reducing memory usage compared to functions that return full lists or collections.


*   When a generator's yield statement is executed, the function's state (including local variables) is saved. Execution resumes from that point on the next iteration, unlike regular functions which lose all state after returning.



*   Generators are especially useful for representing infinite sequences (e.g., Fibonacci numbers, data streams) or when you only need part of a sequence, as they do not require precomputing or storing all values.
*  By using yield, generator functions automatically implement the iterator protocol, removing the need to manually define __iter__() and __next__() methods.

10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?
**Map():-**


*   Purpose: Applies a given function to every item in one or more iterables and returns a new iterable (map object) with the results.
*   Return Value: An iterator containing the results of applying the function to each element.


*   Use Case: When you want to transform or modify all elements in a sequence.
Example-

numbers = [1, 2, 3, 4, 5]

squared = list(map(lambda x: x ** 2, numbers))  # [1, 4, 9, 16, 25]

**Filter():-**

*   Purpose: Filters elements from an iterable by applying a function that returns True or False for each item, keeping only those for which the function returns True.


*   Return Value: An iterator containing only the elements that satisfy the condition.
*   Use Case: When you want to extract elements that meet a certain condition.
Example-

numbers = [1, 2, 3, 4, 5, 6]

even_numbers = list(filter(lambda x: x % 2 == 0, numbers))  # [2, 4, 6]

**Reduce():-**


*   Purpose: Applies a function cumulatively to the items of an iterable, reducing the iterable to a single value.
*   Return Value: A single accumulated result.


*   Use Case: When you need to combine all elements into a single result (such as sum, product, maximum, etc.).

Example-

from functools import reduce

numbers = [1, 2, 3, 4, 5]

product = reduce(lambda x, y: x * y, numbers)  # 120

1.Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in
the list.

```python
def sum_of_even_numbers(numbers):
    even_sum = 0
    for num in numbers:
        if num % 2 == 0:
            even_sum += num
    return even_sum
my_list = [1, 2, 3, 4, 5, 6]
result = sum_of_even_numbers(my_list)
print("Sum of even numbers:", result)

```

4.Write a Python function that checks if a given number is prime or not from 1 to 200.

```python
def is_prime(n):
    if n <= 1:
        return False  # 0 and 1 are not prime numbers
    if n == 2:
        return True   # 2 is the only even prime number
    if n % 2 == 0:
        return False  # All other even numbers are not prime
    for i in range(3, int(n ** 0.5) + 1, 2):
        if n % i == 0:
            return False
    return True

for num in range(1, 201):
    if is_prime(num):
        print(f"{num} is a prime number")

```

5.Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
terms.

```python
class FibonacciIterator:
    def __init__(self, n_terms):
        self.n_terms = n_terms  # Total number of terms to generate
        self.count = 0          # Counter for generated terms
        self.a, self.b = 0, 1   # Starting values for Fibonacci sequence

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.n_terms:
            raise StopIteration
        if self.count == 0:
            self.count += 1
            return 0
        elif self.count == 1:
            self.count += 1
            return 1
        else:
            next_value = self.a + self.b
            self.a, self.b = self.b, next_value
            self.count += 1
            return self.a
n = 10
fib_iter = FibonacciIterator(n)
for num in fib_iter:
    print(num, end=' ')
```

6.Write a generator function in Python that yields the powers of 2 up to a given exponent.

```python
def powers_of_two(max_exponent):
    for exponent in range(max_exponent + 1):
        yield 2 ** exponent

for value in powers_of_two(5):
    print(value)
```

7.Implement a generator function that reads a file line by line and yields each line as a string.

```python
def read_file_lines(filename):
    with open(filename, 'r') as file:
        for line in file:
            yield line.rstrip('\n')  # Removes the trailing newline character

# Example usage:
# Suppose 'example.txt' is a text file in the same directory.
for line in read_file_lines('example.txt'):
    print(line)
```

10.Create a Python program that uses `filter()` to remove all the vowels from a given string.

```python
def remove_vowels(input_str):
    vowels = 'aeiouAEIOU'
    return ''.join(filter(lambda char: char not in vowels, input_str))

text = "I am a student of Data Analysis course"
result = remove_vowels(text)
print("Original String:", text)
print("Without Vowels:", result)
```
